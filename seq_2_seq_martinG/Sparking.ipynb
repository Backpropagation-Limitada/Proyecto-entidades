```python

class sparking():
    
    """
    
    sparking contiene funciones de busqueda de datos en ficheron con texto ,
    con funciones para iniciar cerrar y busuqedas personalizadas de texto
    
    """
    
    def __init__(self,html):
        self.html = html
        self.penta = []
        self.theta = []
        self.octa = []
        self.hepta = []
        
    def iniciar(self,imprimir=False):
        
        try:
            spark=self.html
            self.html=open(spark,'r')
            self.html=self.html.read()
            if imprimir:
                
                return self.html
        except:
            
            if not imprimir:
                return 'setear imprimir para ver los datos'
            else:
                return self.html
    def limpiar(self,limpiar=''):
        
        limpieza=['penta','theta','octa','hepta']
        if limpiar in limpieza:
            if limpiar is limpieza[0]:
                self.penta.clear()
            if limpiar is limpieza[1]:
                self.theta.clear()
            if limpiar is limpieza[2]:
                self.octa.clear()
            if limpiar is limpieza[3]:
                self.hepta.clear()
        if not limpiar:
            self.penta.clear()
            self.theta.clear()
            self.octa.clear()
            self.hepta.clear()
        
    def buscar_datos(self,busqueda='',extra1='',extra2='',limpiar=''):
        #G PRIMARIA DEL DOC
        G=self.html.split(";")
        
        #LIMPIEZA DE DATOS
        
        if limpiar:
            limpieza=['penta','theta','octa','hepta']
            if limpiar in limpieza:
                if limpiar is limpieza[0]:
                    self.penta.clear()
                if limpiar is limpieza[1]:
                    self.theta.clear()
                if limpiar is limpieza[2]:
                    self.octa.clear()
                if limpiar is limpieza[3]:
                    self.hepta.clear()
        
        #BUSQUEDA N 1
        
        if busqueda and not extra1:
            if not self.theta:
                for i in range(len(G)):
                    if busqueda in G[i]:
                        self.theta.extend(G[i].replace("\\n"," ").replace("\t",'').replace("&quot;",'"')\
                                          .replace("&quot",'"').replace('\n',' ').replace("\n\"",'').split("/>"))
            else:
                return 'busqueda completa utilice {}.theta para acceder a los datos'.format('instancia')
        
        #BUSQUEDA N2
        
        if extra1 and not extra2:
            if not self.penta:
                    
#                 for i in range(len(self.theta[0])):
#                     if extra1 in self.theta[0][i]:
#                         self.penta.extend(self.theta[0][i].split(";"))
                    
                for i in range(len(self.theta)):
                    if extra1 in self.theta[i]:
                        self.penta.extend(self.theta[i].split(" "))
            else:
                return 'busqueda con extra1 completa utilice {}.penta para acceder a los datos'.format('instancia')
        
        #BUSQUEDA N3
        
        if extra2:
            if not self.octa:
                
                for i in range(len(self.penta)):
                    if extra2 in self.penta[i]:
                        self.octa.append(self.penta[i].replace("\\n"," ").replace("\t",'').replace("&quot;",'"')\
                                          .replace("&quot",'"').replace('\n',' ').replace("\\"," ").split(" "))
            else:
                return 'utilice {} .octa para acceder a los datos'
        
        #MOSTRAR DATOS
                    
        if busqueda and not extra1:
            return self.theta
        if extra1:
            return self.penta
        if extra2:
            return self.octa
        if not busqueda and not extra1 and not extra2:
            return G
        
    def terminator(self,palabra=''):
        import re
        patrones = re.compile('(title|h1|h2|h3|p|a|link|span|head|body|)')
        
    def cerrar(self):
        try:
            del self.html
        except:
            return 'el objeto fue eliminado'
```
